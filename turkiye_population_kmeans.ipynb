```python
import pandas as pd
```

```python
nufus_verisi=pd.read_csv("nufus_veri.csv")
nufus_verisi
```

```python
import matplotlib.pyplot as plt

plt.scatter(nufus_verisi['yil'],nufus_verisi['nufus'])
plt.xlabel('yil')
plt.ylabel('nufus')
plt.title('nufus_yil_grafigi')
plt.show()
```

```python
from sklearn.cluster import KMeans

kmeans=KMeans(n_clusters=5)
kmeans.fit(nufus_verisi[['yil','nufus']])
nufus_verisi['küme']=kmeans.predict(nufus_verisi[['yil','nufus']])
nufus_verisi
```

```python
plt.scatter(nufus_verisi['yil'],nufus_verisi['nufus'],c=nufus_verisi['küme'])
plt.xlabel('yil')
plt.ylabel('nufus')
plt.title('yila göre nufus')
plt.show()
```

```python
# draw centroids
plt.scatter(nufus_verisi['yil'], nufus_verisi['nufus'], c=nufus_verisi['küme'])
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], c='red')
plt.xlabel('yil')
plt.ylabel('nufus')
plt.title('yila_göre_nufus')
plt.show()
```

```python
nufus_verisi['YeniOlmaDurumu'] = nufus_verisi['küme'].replace({0: 'x_artis', 1: 'x1_artis', 2: 'x2_artis', 3: 'x3_artis', 4: 'x4_artis'})
nufus_verisi
```
